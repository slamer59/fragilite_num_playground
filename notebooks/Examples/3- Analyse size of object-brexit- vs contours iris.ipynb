```python
import pandas as pd
import geopandas as gpd
import geoviews as gv
import sys
gv.extension('bokeh')
```

## Declaring data

```python
geometries = gpd.read_file('../geoviews-examples/assets/boundaries/boundaries.shp')
referendum = pd.read_csv('../geoviews-examples/assets/referendum.csv')
gdf = gpd.GeoDataFrame(pd.merge(geometries, referendum))

gdf.loc[3:6]
```

```python
gdf.loc[5].geometry
```

# Get size

```python
'Size of all geodata: {}'.format(sys.getsizeof(gdf)*1e-6)
```

## Plot

```python
gv.Polygons(gdf.loc[5:], vdims=['name', 'leaveVoteshare'], label='Brexit Referendum Vote').opts(
    tools=['hover'], width=550, height=700, color_index='leaveVoteshare',
    colorbar=True, toolbar='above', xaxis=None, yaxis=None, padding=0.1)
```
